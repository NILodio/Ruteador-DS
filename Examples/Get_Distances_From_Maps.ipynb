```python
def pairwise(iterable):
    a,b = tee(iterable)
    next(b, None)
    return zip(a,b)
```

```python
Distance_value = [0]
Distance_text = [0]

Duration_value =[0]
Duration_text = [0]


# Loop through each row in the data frame using pairwise
for (i1, row1), (i2, row2) in pairwise(df.iterrows()):
    
      #Assign latitude and longitude as origin/departure points
    
    LatOrigin = df['Latitud'].iloc[0]
    LongOrigin = df['Longitud'].iloc[0]
    origins = (LatOrigin,LongOrigin)
    
      #Assign latitude and longitude from the next row as the destination point
    LatDest = row2['Latitud']   
    LongDest = row2['Longitud'] 
    destination = (LatDest,LongDest)
        
    result_get = gmaps.distance_matrix(origins, destination, mode='driving')
    
    
    #append result to list
    Distance_value.append(result_get['rows'][0]['elements'][0]['distance']['value'])
    Distance_text.append(result_get['rows'][0]['elements'][0]['distance']['text'])
    
    Duration_value.append(result_get['rows'][0]['elements'][0]['duration']['value'])
    Duration_text.append(result_get['rows'][0]['elements'][0]['duration']['text'])
    
    
```

```python
df['Distancia'] = Distance_value
df['Distancia_text'] = Distance_text
df['Duración'] = Duration_value
df['Duración_text'] = Duration_text
```
